```python
'''LLoyd Algorithm'''

def EuclidianDistance(pointA,Centers):
        import math
        ''' find the euclidian distance between point A and B in a m dimensional space
        INPUT:
        pointA=[0,0]
        Centers=[[5.0,5.0],[0.0,5.0],[1.0,1.0],[2.0,2.0],[3.0,3.0],[1.0,2.0]]
        OUTPUT= 1.41421356237
        '''
        return min([math.sqrt(sum([(pointA[m]-pointB[m])**2 for m in range(len(pointA))]))for pointB in Centers])
    
def CenterOfGravity(Points):
    return [sum([point[m] for point in Points])/len(Points) for m in range(len(Points[0]))]
    
###################

#f=open('input.txt','r')
f=open('../../Downloads/rosalind_ba8c.txt','r')
lines=f.read().splitlines()
f.close()

#print lines[0]
k=int(lines[0].split(' ')[0])
m=lines[0].split(' ')[1]
DataPoints=[map(float,[item for item in line.split(' ')]) for line in lines[1:]]

#init
#print 'Centers',Centers
def Lloyd_algorithm(DataPoints):
    
    def EuclidianDistance(pointA,Centers):
        import math
        ''' find the euclidian distance between point A and B in a m dimensional space
        INPUT:
        pointA=[0,0]
        Centers=[[5.0,5.0],[0.0,5.0],[1.0,1.0],[2.0,2.0],[3.0,3.0],[1.0,2.0]]
        OUTPUT= 1.41421356237
        '''
        return min([math.sqrt(sum([(pointA[m]-pointB[m])**2 for m in range(len(pointA))]))for pointB in Centers])
    
    def CenterOfGravity(Points):
        return [sum([point[m] for point in Points])/len(Points) for m in range(len(Points[0]))]

    Centers= DataPoints[:k]
    clouds_old=[]
    while True:
        #centers to clusters
        Center_dic={}
        for point in Centers:
            Center_dic[str(point)]=[]
        for Point in DataPoints:
            mindist=EuclidianDistance(Point,Centers)
            center=[pointB for pointB in Centers if EuclidianDistance(Point,[pointB])==mindist][0]
            Center_dic[str(center)]+=[Point]
        #Clusters to centers
        clouds_new=[]
        for point in Centers:
            cloud=Center_dic[str(point)]
            clouds_new+=[cloud]
            gravity= CenterOfGravity(cloud)
            Center_dic[str(gravity)] = Center_dic.pop(str(point))
        Centers=[map(float,point[1:-1].split(', ')) for point in Center_dic.keys()]
        if clouds_new != clouds_old:
            clouds_old=clouds_new
        else:
            return Centers
            break
        
    
print '\n'.join([' '.join(map(str,[coord for coord in item])) for item in Lloyd_algorithm(DataPoints)])

   
```

```python
def UPGMA(D,n):
    
    '''
    INPUT:
    n = 4
    matrix = [[ 0 20 17 11]
     [20  0 20 13]
     [17 20  0 10]
     [11 13 10  0]]
     OUTPUT=defaultdict(<type 'list'>, {0: [[5, 7.0]], 1: [[6, 8.833333333333334]], 
     2: [[4, 5.0]], 3: [[4, 5.0]], 4: [[3, 5.0], [2, 5.0], [5, 2.0]], 
     5: [[4, 2.0], [0, 7.0], [6, 1.833333333333334]], 
     6: [[5, 1.833333333333334], [1, 8.833333333333334]]} # where dic[key] = node source
                                                          # dic value[0]= node sink
                                                          # dic value[1]= branch length
    '''
    import pandas as pd
    import numpy as np
    from collections import defaultdict
    Clusters=map(str,range(n+1)[1:])
    D[D == 0] = float('Nan')
    D = pd.DataFrame(D, index=Clusters, columns=Clusters)
    T=[]
    leaf_counter={}
    for node in Clusters:
        leaf_counter[node]=1
    while len(Clusters)>1:
        min_val=np.nanmin(D) #D.values.min()
        for x in Clusters:
            for y in Clusters:
                D[x][y]
                if D[x][y]==min_val:
                    C1,C2=x,y
                    continue
        Cnew=str(C1)+', '+str(C2)
        T+=[Cnew]
        leaf_counter[str(Cnew)]=leaf_counter[C1]+leaf_counter[C2]
        D[str(Cnew)]=(D[C1]*leaf_counter[C1]+D[C2]*leaf_counter[C2])/leaf_counter[str(Cnew)]
        D=D.append(D[str(Cnew)], ignore_index=False)
        D[Cnew][Cnew]=float('NaN')
        del D[C1]
        del D[C2]
        D=D[D.index != C1]
        D=D[D.index != C2]
        Clusters.remove(C1)
        Clusters.remove(C2)
        Clusters.append(Cnew)
    root=Clusters[0]

    return T



#f=open('input.txt','r')
#f=open('input2.txt','r')
f=open('../../Downloads/rosalind_ba8e.txt','r')
lines=f.read().splitlines()
f.close()
n=int(lines[0])
matrix=[map(float,line.split()) for line in lines[1:] ]
matrix=np.matrix(matrix)
matrix=matrix[:n,:n]
#print matrix

print '\n'.join([' '.join(item.split(', ')) for item in UPGMA(matrix,n)])
```
