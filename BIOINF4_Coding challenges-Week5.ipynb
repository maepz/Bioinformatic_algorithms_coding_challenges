```python
'''CODE CHALLENGE: Solve the Peptide Identification Problem.'''
    
def PeptideIdentification(Spectral_v,Proteome,prot):

    '''
    Find a peptide from a proteome with maximum score against a spectrum
    !!! Add [0] to Spectral Vector
    Spectral Vector= [0,0,0,4,-2,-3,-1,-7,6,5,3,2,1,9,3,-8,0,3,1,2,1,8] #A space-delimited spectral vector Spectrum'
    INPUT:
    Spectral_v=[0,0,0,0,4,-2,-3,-1,-7,6,5,3,2,1,9,3,-8,0,3,1,2,1,8]
    Proteome=XZZXZXXXZXZZXZXXZ # An amino acid string Proteome.
    prot={'X':4,'Z':5} # aa <-> mass dictionary
    OUTPUT= ZXZXX # A substring of Proteome with maximum score against Spectrum'.
    '''    
    
    def ScorePeptideAgainstSpectralVector(Peptide,Spectral_v,prot):
        index=0
        score=0
        max_pep_weight=len(Spectral_v)-1
        for n in range(len(Peptide)):
            aa=Peptide[n]
            index+=prot[aa]
            if index>max_pep_weight:
                return 'Peptide too massive'
            else:
                score+=Spectral_v[index]
        if index<max_pep_weight:
             return 'Peptide too light'
            
        return score

    #############################################################################
    pep_score=[]
    max_score=-float('Inf')
    max_pep=''
    max_pep_len=len(Spectral_v)-1
    for n in range(0,len(Proteome)):
        for i in range(n+1,len(Proteome)):
            peptide= Proteome[n:i]
            if peptide in pep_score:
                pass
            else:
                pep_score+=[peptide]
                score=ScorePeptideAgainstSpectralVector(peptide,Spectral_v,prot)
                if score == 'Peptide too massive':
                    break
                if score == 'Peptide too light':
                    pass
                elif score>max_score:
                    max_score=score
                    max_pep=peptide
    return max_pep
    



#spec=[0,0,0,4,-2,-3,-1,-7,6,5,3,2,1,9,3,-8,0,3,1,2,1,8]
spec=[4, -3, -2, 3, 3, -4, 5, -3, -1, -1, 3, 4, 1, 3]
Proteome='XZZXZXXXZXZZXZXXZ'
prot={'X':4,'Z':5}

'''
f=open('input.txt','r')
#f=open('../../Downloads/dataset_11866_2(4).txt','r')
#f=open('../../Downloads/rosalind_ba11f.txt','r')
lines=f.read().splitlines()
f.close()
spec=map(int,lines[0].split())
Proteome=lines[1]
#print Proteome
#print spec
#prot={'A': 71, 'C': 103, 'E': 129, 'D': 115, 'G': 57, 'F': 147, 'H': 137, 'M': 131, 'I':113, 'L': 113, 'N': 114, 'K':128, 'Q': 128, 'P': 97, 'S': 87, 'R': 156, 'T': 101, 'W': 186, 'V': 99, 'Y': 163}
'''

spec=[0]+spec
#print spec
#print ScorePeptideAgainstSpectralVector('VEAAFYAKITNHSGH',spec,prot)
print PeptideIdentification(spec,Proteome,prot)
```

```python
'''PSMSearch(SpectralVectors, Proteome, threshold).
    PSMSet ← an empty set
    for each vector Spectrum' in SpectralVectors
          Peptide ← PeptideIdentification(Spectrum', Proteome)
          if Score(Peptide, Spectrum) ≥ threshold
              add the PSM (Peptide, Spectrum') to PSMSet
    return PSMSet'''
def PSMsearch(SpectralVectors, Proteome, threshold,prot=[]):
    
    ''' 
    Identify Peptide-Spectrum Matches by matching spectra against a proteome.
    INPUT:
    SpectralVectors = [[-1, 5, -4, 5, 3, -1, -4, 5, -1, 0, 0, 4, -1, 0, 1, 4, 4, 4], [-4, 2, -2, -4, 4, -5, -1, 4, -1, 2, 5, -3, -1, 3, 2, -3]] # A list of spectral vectors 
    Proteome = XXXZXZXXZXZXXXZXXZX # an amino acid string Proteome
    threshold = 5 # a integer score threshold
    prot = {'X':4,'Z':5} # aa <-> mass dictionary
    OUTPUT = ['XZXZ'] # A list of  all unique Peptide-Spectrum Matches scoring at least as high threshold.
    '''
 
    def ScorePeptideAgainstSpectralVector(Peptide,Spectral_v,prot):
        index=0
        score=0
        max_pep_weight=len(Spectral_v)-1
        for n in range(len(Peptide)):
            aa=Peptide[n]
            index+=prot[aa]
            if index>max_pep_weight:
                return 'Peptide too massive'
            else:
                score+=Spectral_v[index]
        if index<max_pep_weight:
             return 'Peptide too light'
        return score
    
    def PeptideIdentification2(Spectral_v,Proteome,prot):
        '''
        Find a peptide from a proteome with maximum score against a spectrum. To run faster, this version first evaluate the min size of peptide and 
        skip peptides too small
        !!! Add [0] to Spectral Vector
        Spectral Vector= [0,0,0,4,-2,-3,-1,-7,6,5,3,2,1,9,3,-8,0,3,1,2,1,8] #A list of spectral vector Spectrum'
        INPUT:
        Spectral_v=[0,0,0,0,4,-2,-3,-1,-7,6,5,3,2,1,9,3,-8,0,3,1,2,1,8]
        Proteome=XZZXZXXXZXZZXZXXZ # An amino acid string Proteome.
        prot={'X':4,'Z':5} # aa <-> mass dictionary
        OUTPUT= ZXZXX # A substring of Proteome with maximum score against Spectrum'.
        '''    
        pep_score=[]
        max_score=-float('Inf')
        max_pep=''
        max_pep_weight=len(Spectral_v)-1
        max_aa=max(prot.values())
        if max_pep_weight%max_aa>0:
            min_pep_len=max_pep_weight/max_aa+1
        else:
            min_pep_len=max_pep_weight/max_aa
        for n in range(0,len(Proteome)):
            for i in range(n+min_pep_len,len(Proteome)):
                peptide= Proteome[n:i]
                if peptide in pep_score:
                    pass
                else:
                    pep_score+=[peptide]
                    score=ScorePeptideAgainstSpectralVector(peptide,Spectral_v,prot)
                    if score == 'Peptide too massive':
                        break
                    if score == 'Peptide too light':
                        pass
                    elif score>max_score:
                        max_score=score
                        max_pep=peptide
        return max_pep,max_score
    ###############################################################################
    prot=prot
    PSMset=[]
    for Spectrum_v in SpectralVectors:
        Spectrum_v=[0]+Spectrum_v #prepond spectral vector with 0
        Peptide,score = PeptideIdentification2(Spectrum_v,Proteome,prot)
        if score>=threshold:
            PSMset+=[Peptide]
    return PSMset
        
############

f=open('input.txt','r')
#f=open('input2.txt','r')
#f=open('../../Downloads/dataset_11866_5.txt','r')
#f=open('../../Downloads/rosalind_ba11g.txt','r')
lines=f.read().splitlines()
f.close()

SpectralVectors=[]
for n in range(len(lines)):
    line=lines[n]
    try:
        if len(map(int,line.split()))>1:
            SpectralVectors+=[map(int,line.split())]
        else:
            threshold=int(line)
    except ValueError:
        Proteome=line
    
print SpectralVectors
print Proteome
print threshold
#prot={'A': 71, 'C': 103, 'E': 129, 'D': 115, 'G': 57, 'F': 147, 'H': 137, 'M': 131, 'I':113, 'L': 113, 'N': 114, 'K':128, 'Q': 128, 'P': 97, 'S': 87, 'R': 156, 'T': 101, 'W': 186, 'V': 99, 'Y': 163}
prot={'X':4,'Z':5}


        
print '\n'.join(PSMsearch(SpectralVectors, Proteome, threshold,prot=prot))
```

```python
'''CODE CHALLENGE: Solve the Size of Spectral Dictionary Problem.
     Given: A spectral vector Spectrum', an integer threshold, and an integer max_score.
     Return: The size of the dictionary Dictionarythreshold(Spectrum').

'''
import time
start=time.time()

#f=open('input.txt','r')
#f=open('input2.txt','r')
#f=open('../../Downloads/dataset_11866_8.txt','r')
f=open('../../Downloads/rosalind_ba11h.txt','r')
lines=f.read().splitlines()
f.close()

Spectrum=[0]+map(int,lines[0].split())
#print Spectrum
threshold=int(lines[1])
max_score=int(lines[2])
#print len(Spectrum)-1
#print threshold
#print max_score
prot={'A': 71, 'C': 103, 'E': 129, 'D': 115, 'G': 57, 'F': 147, 'H': 137, 'M': 131, 'I':113, 'L': 113, 'N': 114, 'K':128, 'Q': 128, 'P': 97, 'S': 87, 'R': 156, 'T': 101, 'W': 186, 'V': 99, 'Y': 163}
#prot={'X':4,'Z':5}

def DicThresholdSpec(Spectrum, threshold, max_score,prot): 
    '''
    Find the number of peptides that match Spectrum with a score between threshold and max_score.
    Uses 'prot' aa masse table. 
    INPUT:
    Spectrum=[0, 4, -3, -2, 3, 3, -4, 5, -3, -1, -1, 3, 4, 1, 3]
    threshold=1
    max_score=8
    prot={'X':4,'Z':5}
    OUTPUT = 3
    '''
    import numpy as np
    m=len(Spectrum)-1
    size = np.zeros((m+1,max_score+1),dtype=int)
    size[0,0]=1
    for i in range(1,m+1):
        for t in range(max_score+1): 
            size[i,t] = sum([size[i-aa_weight,t-Spectrum[i]] for aa_weight in prot.values() if (i-aa_weight)>=0 and (t-Spectrum[i])>=0 and (t-Spectrum[i])<=max_score])
    result=sum(size[m,t] for t in range(threshold,max_score+1))        
    return result


size= DicThresholdSpec(Spectrum, threshold, max_score,prot)
print size
print 'Time it took',time.time()-start,'s'
```

```python
'''CODE CHALLENGE: Solve the Probability of Spectral Dictionary Problem.'''

import time
start=time.time()

#f=open('input.txt','r')
#f=open('input2.txt','r')
#f=open('../../Downloads/dataset_11866_11.txt','r')
f=open('../../Downloads/rosalind_ba11i.txt','r')
lines=f.read().splitlines()
f.close()

Spectrum=[0]+map(int,lines[0].split())
#print Spectrum
threshold=int(lines[1])
max_score=int(lines[2])
#print len(Spectrum)-1
#print threshold
#print max_score
prot={'A': 71, 'C': 103, 'E': 129, 'D': 115, 'G': 57, 'F': 147, 'H': 137, 'M': 131, 'I':113, 'L': 113, 'N': 114, 'K':128, 'Q': 128, 'P': 97, 'S': 87, 'R': 156, 'T': 101, 'W': 186, 'V': 99, 'Y': 163}
#prot={'X':4,'Z':5}

def Pr_DicThresholdSpec(Spectrum, threshold, max_score,prot): 
    '''
    Find the probability for a peptide match against Spectrum with a score between threshold and max_score.
    Find the probability of the spectral dictionary for a given spectrum and score threshold.
    Uses 'prot' aa masse table. 
    INPUT:
    Spectrum=[0, 4, -3, -2, 3, 3, -4, 5, -3, -1, -1, 3, 4, 1, 3]
    threshold=1
    max_score=8
    prot={'X':4,'Z':5}
    OUTPUT = 0.375
    '''
    import numpy as np
    m=len(Spectrum)-1
    size = np.zeros((m+1,max_score+1),dtype=float)
    size[0,0]=1
    dimensions=float(1)/len(prot) # can alos be written 1./len(prot)
    for i in range(1,m+1):
        for t in range(max_score+1): 
            size[i,t] = dimensions * sum([size[i-aa_weight,t-Spectrum[i]] for aa_weight in prot.values() if (i-aa_weight)>=0 and (t-Spectrum[i])>=0 and (t-Spectrum[i])<=max_score])
    result=sum(size[m,t] for t in range(threshold,max_score+1))        
    return result


size= Pr_DicThresholdSpec(Spectrum, threshold, max_score,prot)
print size
print 'Time it took',time.time()-start,'s'
```

```python
'''CODE CHALLENGE 5: Solve the Spectral Alignment Problem.'''
import numpy as np
import time
import pandas as pd

def PeptideToVector2(peptide,prot=[]):
    
    '''
    Convert a peptide into a binary peptide vector.
    INPUT = XZZXX # A peptide P.
    OUTPUT = 0001000010000100010001 # The peptide vector of P.
    prot={'X':4,'Z':5}
    '''
    prot=prot
    vector_len=sum([prot[n] for n in peptide])
    vector=[0]
    for n in range(1,len(peptide)+1):
        mass=sum([prot[n] for n in peptide[:n]])
        vector+=[mass]
    return vector


Peptide='XXZ'
Spectrum=[0, 4, -3, -2, 3, 3, -4, 5, -3, -1, -1, 3, 4, 1, -1]
k=2


prot={'X':4,'Z':5}
#prot={'A': 71, 'C': 103, 'E': 129, 'D': 115, 'G': 57, 'F': 147, 'H': 137, 'M': 131, 'I':113, 'L': 113, 'N': 114, 'K':128, 'Q': 128, 'P': 97, 'S': 87, 'R': 156, 'T': 101, 'W': 186, 'V': 99, 'Y': 163}

'''
#f=open('input.txt','r')
#f=open('input2.txt','r')
#f=open('../../Downloads/dataset_11866_14.txt','r')
f=open('../../Downloads/rosalind_ba11j.txt','r')
lines=f.read().splitlines()
f.close()
Peptide=lines[0]
Spectrum=[0]+map(int,lines[1].split(' '))
k=int(lines[2])
'''
start=time.time()

print Peptide
print PeptideToVector2(Peptide,prot=prot)
print len(Spectrum)

def SpectralAlignment(Peptide,Spectrum,k,prot):
    '''
    Given a peptide and a spectral vector, find a modified variant of this peptide that maximizes the peptide-spectrum score
    among all variants of the peptides with up to k modifications.
    
    INPUT:
    Peptide='XXZ'
    Spectrum=[0, 4, -3, -2, 3, 3, -4, 5, -3, -1, -1, 3, 4, 1, -1]
    k=2
    prot={'X':4,'Z':5}
    
    OUTPUT= 
    
    '''

    ####### create dictionnary of dataframes and diff dictionnary ############
    
    pep_vector=[0]
    for n in range(1,len(Peptide)+1):
        mass=sum([prot[n] for n in Peptide[:n]])
        pep_vector+=[mass]
        
    
    df=pd.DataFrame(np.zeros((len(pep_vector),len(Spectrum)),dtype=float))
    df.index=pep_vector
    dfdic={}
    for n in range(k+1):
        dfdic[n]=df.copy()

    diff={}
    for n in range(1,len(pep_vector)):
        index=pep_vector[n]
        diff[index]=pep_vector[n]-pep_vector[n-1]

    ###### initiate tables and backtrack ############

    backtrack={}
    backtrack['[0, 0, 0]']=[[0,0,0]]
    for t in range(k+1):
        for i in pep_vector:
            for j in range(len(Spectrum)):
                dfdic[t].loc[i,j]=-float('Inf')
                if j==0 or i==0:
                    coord=[t,i,j]
        dfdic[t].loc[0,0]=0    

    ######### fill in first table t=0 #######
    t=0
    for i in pep_vector[1:]:
        for j in [vec for vec in pep_vector[1:] if vec in range(len(Spectrum))]:
            coord=[t,i,j]
            #print coord

            if j-diff[i]>=0:

                score_tpred=dfdic[t].loc[i-diff[i],j-diff[i]]
                dfdic[t].loc[i,j]=Spectrum[j]+score_tpred

                predec=[t,i-diff[i],j-diff[i]]

                try:
                    backtrack[str(coord)]=backtrack[str(predec)]+[coord]
                except KeyError:
                    pass

            else:
                pass

    ####### fill-in tables t>0 #############

    for t in range(1,k+1):
        for i in pep_vector[1:]:
            for j in range(1,len(Spectrum)):
                coord=[t,i,j]
                idx_pred=[]
                if j-diff[i]<0:
                    score_tpred=-float('Inf')
                else:    
                    score_tpred=dfdic[t].loc[i-diff[i],j-diff[i]]
                score_opred=list(dfdic[t-1].loc[i-diff[i]][:j])
                all_scores=[score_tpred]+score_opred

                score=max(all_scores)
                if score !=-float('Inf'):
                    dfdic[t].loc[i,j]=Spectrum[j]+score

                    # in case of multiple max_scores, discard the ones that don't lead to source (0,0,0)
                    max_scores=[a for a, b in enumerate(all_scores) if b == score]
                    if len(max_scores)>1: 
                        for n in max_scores:
                            if n==0:
                                predec=[t,i-diff[i],j-diff[i]]
                                try:
                                    backtrack[str(predec)]
                                    break
                                except KeyError:
                                    pass
                            else:
                                predec=[t-1,i-diff[i],n-1]
                                try:
                                    backtrack[str(predec)]
                                    break
                                except KeyError:
                                    pass
                    else:
                        if score == score_tpred:
                            predec=[t,i-diff[i],j-diff[i]]
                        else:
                            predec=[t-1,i-diff[i],score_opred.index(max(score_opred))]

                try:
                    backtrack[str(coord)]=backtrack[str(predec)]+[coord]
                except KeyError:
                    pass

    ####### find max value and format result ############

    max_score=max([dfdic[t].loc[pep_vector[-1],len(Spectrum)-1] for t in range(k+1)])
    t_idx_max=[t for t in range(k+1) if dfdic[t].loc[pep_vector[-1],len(Spectrum)-1]==max_score]
    if len(t_idx_max)==1:
        coord_max=[t_idx_max[0],pep_vector[-1],len(Spectrum)-1]
    else:
        for item in t_idx_max:
            coord_max=[t_idx_max,pep_vector[-1],len(Spectrum)-1]
            try:
                backtrack[str(coord_max)]
            except KeyError:
                pass

    modif_pep_vector=[j for t,i,j in backtrack[str(coord_max)]]

    result=''
    for n in range(1,len(pep_vector)):
        aa=Peptide[n-1]    
        delta_weight=modif_pep_vector[n]-modif_pep_vector[n-1]-prot[aa]
        if delta_weight==0:
            result+=aa
        else:
            result+=aa+'('+'%+d' % delta_weight+')'
    return result

print SpectralAlignment(Peptide,Spectrum,k,prot)

print 'the script took ',time.time()-start,' s.'
```
