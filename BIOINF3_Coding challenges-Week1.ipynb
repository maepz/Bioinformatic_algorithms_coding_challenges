```python
''' FIND longest path in DAG

pseudocode: 
 for each node b in Graph
            sb ← −∞
        ssource ← 0
        topologically order Graph
        for each node b in Graph (following the topological order)
            sb ← maxall predecessors a of node b {sa + weight of edge from a to b}
        return ssink
'''


class  DAG:
    def __init__(self, x):
        self.graph = x
        self.description = "a directed acyclic graph (DAG) with weight of edges. Example:'0->1:7','0->2:4','2->3:2','1->4:1','3->4:3'"

    def weight(self):
        '''dictionary of all path weights'''
        weight_dic={}
        for line in self.graph:
            source=line.split('->')[0]
            sink=(line.split('->')[1]).split(':')[0]
            weight=(line.split('->')[1]).split(':')[1]
            weight_dic[source+'->'+sink]=weight
        return weight_dic
    
    def successor(self):
        '''dictionary of all node sucessors'''
        from collections import defaultdict
        dic_successor=defaultdict(list)
        for line in self.graph:
            source=line.split('->')[0]
            sink=(line.split('->')[1]).split(':')[0]
            if source in dict(dic_successor).keys():
                dic_successor[source].append(sink)
            else: 
                dic_successor[source].append(sink)
        return dict(dic_successor)      
            
    def predecessor(self):
        '''dictionary of all node predecessors'''
        from collections import defaultdict
        dic_predecessor=defaultdict(list)
        for line in self.graph:
            source=line.split('->')[0]
            sink=(line.split('->')[1]).split(':')[0]
            if sink in dict(dic_predecessor).keys():
                dic_predecessor[sink].append(source)
            else:
                dic_predecessor[sink].append(source)
        return dict(dic_predecessor)
    
    def clean_graph(self,source,sink):
        ''' iteratively remove all paths in DAG from starting points that are not the source '''
        graph=self.graph
        startingpoints=[key for key in self.successor().keys() if key not in self.predecessor().keys()]
        while True:
            if startingpoints==[str(source)]:
                break
            for line in graph[:]:
                if line.split('->')[0]==str(source):
                    pass
                elif line.split('->')[0] in startingpoints:
                    graph.remove(line)    
            newdag=DAG(graph)
            graph=newdag.graph
            startingpoints=[key for key in newdag.successor().keys() if key not in newdag.predecessor().keys()]
        return graph
    
    def topolgical_sort(self):
        ''' sort DAG topologically'''
        graph_sorted = []
        graph_unsorted=self.successor().copy()
        while graph_unsorted:
            acyclic = False
            for node, edges in graph_unsorted.items():
                for edge in edges:
                    if edge in graph_unsorted:
                        break
                else:
                    acyclic = True
                    del graph_unsorted[node]
                    graph_sorted.append((node, edges))
            if not acyclic:
                raise RuntimeError("A cyclic dependency occurred")

        return graph_sorted[::-1]
      
    def topological_nodes(self):
        ''' provide list of nodes ordered according to topological order in DAG'''
        top= self.topolgical_sort()[:]
        node_list=[]
        for path in top:
            source=path[0]
            if source in node_list:
                continue
            else:
                    node_list.append(source)
        for path in top:
            for sink in path[1]:
                node=sink
                if node in node_list:
                    continue
                elif node==str(0):
                    continue
                else:
                        node_list.append(node)
        return node_list
    
    def LongestPath(self, source, sink):
        ''' finds longest path from source to sink in DAG'''
        s={}
        path={}
        s[str(source)]=0
        path[str(source)]=source
        node_list=self.topological_nodes()[:]
        for node in node_list:
            try:
                s[node]=max([int(s[item])+int(self.weight()[item+'->'+node]) for item in self.predecessor()[node]])
                for item in self.predecessor()[node]:
                    if int(s[item])+int(self.weight()[item+'->'+node]) == s[node]:
                        path[node]=str(path[item])+'->'+node
            except KeyError:
                continue
        return s[str(sink)], path[str(sink)]

#### TEST ####
''''
start=0
end=4
lines='0->1:7','0->2:4','2->3:2','1->4:1','3->4:3'
dag=DAG(lines)
'''
'''    
f=open('input.txt','r')
lines=f.read().splitlines()
start=lines[0]
end=lines[1]
dag=DAG(lines[2:])

print DAG(dag.clean_graph(0,44)).graph
print DAG(dag.clean_graph(0,44)).topolgical_sort()
print DAG(dag.clean_graph(0,44)).topological_nodes()
print DAG(dag.clean_graph(0,44)).LongestPath_weight(0,44)
'''
#### CHALLENGE ####

f=open('../../Downloads/dataset_245_7.txt','r')
f=open('../../Downloads/rosalind_ba5d.txt','r')
lines=f.read().splitlines()
source=lines[0]
sink=lines[1]
dag=DAG(lines[2:])

print '\n'.join(map(str,DAG(dag.clean_graph(source,sink)).LongestPath(source,sink)))
```
