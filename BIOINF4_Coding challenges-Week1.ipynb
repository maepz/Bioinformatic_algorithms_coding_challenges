```python
'''Distance Between Leaves Problem

Compute the distances between leaves in a weighted tree.

Given: An integer n followed by the adjacency list of a weighted tree with n leaves.

Return: A space-separated n x n (di, j), where di, j is the length of the path between leaves i and j.'''


def distances_between_leaves(n,adjacency_list):
    """    
    Distances Between Leaves Problem: Compute the distances between leaves in a weighted tree.
    Input:  An integer n followed by the adjacency list of a weighted tree with n leaves.
    Output: An n x n matrix (di,j), where di,j is the length of the path between leaves i and j.
    """
    import numpy as np
    adj = {}
    for (k, v, w) in adjacency_list:
        adj.setdefault(k,[]).append((v,w))
    #print 'adj',adj    
    # assume the keys are given from [0,....,p]
    #check if assumption is true
    assert min(adj.keys()) == 0 # assert [condition] tells the program to test that condition, and trigger an error if the condition is false
    assert max(adj.keys()) == len(adj.keys()) - 1
    
    #find out leaves
    leaves = [k for k,v in adj.iteritems() if len(v) == 1]
    assert len(leaves) == n
    #print leaves
    
    #make empty arrays with right size
    node_to_node_distance = np.zeros( (len(adj),len(adj)), dtype = int )
    #print 'node_to_node_distance',node_to_node_distance
    leaf_to_leaf_distance = np.zeros( (n,n), dtype = int )   
    #print 'leaf_to_leaf_distance',leaf_to_leaf_distance
    
    
    def dfs(path):
        import numpy as np
        for v,w in adj[path[-1][0]]:
            if visited[v] == True:
                continue
            visited[v] = True
            pathlen = path[-1][1] + w
            for (iv,ilen) in path:
                node_to_node_distance[iv,v] = pathlen - ilen
                node_to_node_distance[v,iv] = pathlen - ilen
            npath = path[:]
            npath.append((v,pathlen))
            dfs(npath)
            
            
    for leaf in leaves:
        visited = [None] * len(adj)
        visited[leaf] = True
        path = [(leaf,0)]
        #print 'path',path
        dfs(path)
        #print 'dfs path',dfs(path)
        #print 'node_to_node_distance',node_to_node_distance
    for i,m in enumerate(leaves):
        for j,n in enumerate(leaves):
            leaf_to_leaf_distance[i,j] = node_to_node_distance[m,n]
    return leaf_to_leaf_distance

#m = distances_between_leaves(4,[(0, 4, 11), (1, 4, 2), (2, 5, 6), (3, 5, 7), (4, 0, 11), (4, 1, 2), (4, 5, 4), (5, 4, 4), (5, 3, 7), (5, 2, 6)])
#res = np.array([[ 0, 13, 21, 22],[13,  0, 12, 13],[21, 12,  0, 13],[22, 13, 13,  0]])
#assert not (m != res).sum()



### test 1 ###


f=open('input.txt','r')
lines=f.read().splitlines()
f.close()
leaves=int(lines[0])
adj=lines[1:]


### test 3 ###

f=open('input.txt','r')
lines=f.read().splitlines()
f.close()
leaves=int(lines[0])
adj=lines[1:]


### challenge ###
'''
f=open('../../Downloads/rosalind_ba7a.txt','r')
lines=f.read().splitlines()
f.close()
leaves=int(lines[0])
adj=lines[1:]
'''
#####
import numpy as np
np.set_printoptions(threshold=np.inf)

adjency_list=[]
for item in adj:
    node=[[item.split('->')[0]]+[item.split('->')[1].split(':')[0]]+[item.split('->')[1].split(':')[1]]]
    node=[map(int,item) for item in node]
    adjency_list+=node

print adjency_list
#distances_between_leaves(leaves,adjency_list)
#print str(distances_between_leaves(leaves,adjency_list)).replace('[','').replace(']','')

#np.savetxt('test.out', distances_between_leaves(leaves,adjency_list), delimiter=' ',fmt='%1i')
```
